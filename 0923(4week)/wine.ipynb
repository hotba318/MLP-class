```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("/content/drive/MyDrive/2025-2 MLP/source/wine+quality/winequality-white.csv", sep=';')
df
```

```python
df.isnull().sum()
```

```python
X = df.drop('quality', axis=1)
y = df['quality']
X, y
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(X_train)
x_test_scaled = scaler.transform(X_test)
```

```python
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=200, random_state=42)
lr = LogisticRegression(max_iter=10000)
knn = KNeighborsClassifier(n_neighbors=5)

dt.fit(x_train_scaled, y_train)
rf.fit(x_train_scaled, y_train)
lr.fit(x_train_scaled, y_train)
knn.fit(x_train_scaled, y_train)

dt_pred = dt.predict(x_test_scaled)
rf_pred = rf.predict(x_test_scaled)
lr_pred = lr.predict(x_test_scaled)
knn_pred = knn.predict(x_test_scaled)

dt_acc = accuracy_score(y_test, dt_pred)
rf_acc = accuracy_score(y_test, rf_pred)
lr_acc = accuracy_score(y_test, lr_pred)
knn_acc = accuracy_score(y_test, knn_pred)

print(f"Decision Tree Accuracy: {dt_acc}")
print(f"Random Forest Accuracy: {rf_acc}")
print(f"Logistic Regression Accuracy: {lr_acc}")
print(f"KNeighbors Accuracy: {knn_acc}")
```

```python
print("Logistic Regression Intercept:", lr.intercept_)
print("Logistic Regression Coefficients:", lr.coef_)
```
